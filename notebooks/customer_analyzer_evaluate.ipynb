```python
import pandas as pd
import joblib
from sklearn.metrics import classification_report, confusion_matrix
from sqlalchemy import create_engine
```

```python
model = joblib.load("logistic_regression.pkl")
tfidf = joblib.load("tfidf_vectorizer.pkl")
```

```python
engine = create_engine("postgresql://postgres:Charvi1234!@localhost:5433/Customer Feedback Analyzer DB")

clean_df = pd.read_sql("SELECT * FROM cleaned_reviews", engine)
```

```python
X = clean_df['cleaned_review']
y = clean_df['sentiment_score']
```

```python
X_tfidf = tfidf.transform(X)
```

```python
y_pred = model.predict(X_tfidf)
```

```python
print(classification_report(y, y_pred))
```

```python
samples = [
    "Great service and super fast delivery!",
    "Terrible experience, my product arrived damaged."
]

samples_tfidf = tfidf.transform(samples)
model.predict(samples_tfidf)
```
